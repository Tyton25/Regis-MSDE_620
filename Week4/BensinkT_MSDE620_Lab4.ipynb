# Introduction
### This notebook cleans a dataset using string operations to replace and substitute variable names and values whose meaning are unclear to anyone performing analysis on the data. After some basic EDA, characters such as '_', '-', and '.' will be replaced with a space. Additionally, column headers will be capitalized and abbreviations will be replaced with full words to maximize clarity.

```python
import re
import pandas as pd
import numpy as np
```

## Exploratory DataFrame Analysis

```python
df = pd.read_csv('assign_wk4/bank-additional.csv', sep=';')
```

```python
df.head()
```

```python
df.tail()
```

## Rename Columns

```python
columns = df.columns
columns
```

### Replace characters using list comprehensions

```python
columns = [c.replace('.', ' ') for c in columns]
columns = [c.replace('_', ' ') for c in columns]
columns = [c.replace('cons ', 'consumer ') for c in columns]
columns = [c.replace(' idx', ' index') for c in columns]
columns = [c.replace(' conf', ' confidence') for c in columns]
columns
```

```python
df.columns = columns
df
```

### Rename individual columns for clarity

```python
df.rename(columns = {'nr employed':'number employed'}, inplace=True)
df.rename(columns = {'poutcome':'previous outcome'}, inplace=True)
df.rename(columns = {'pdays':'days since contacted'}, inplace=True)
df.rename(columns = {'previous':'number previous contacts'}, inplace=True)
df.rename(columns = {'y':'term deposited'}, inplace=True)
df.rename(columns = {'euribor3m':'eurobori 3 mos'}, inplace=True)
df.rename(columns = {'duration':'contact duration'}, inplace=True)
df.rename(columns = {'marital':'marital status'}, inplace=True)
df.rename(columns = {'job':'job type'}, inplace=True)
df.rename(columns = {'housing':'mortgage'}, inplace=True)
df.rename(columns = {'loan':'personal loan'}, inplace=True)
df.rename(columns = {'default':'default credit'}, inplace=True)
df.rename(columns = {'month':'last contact'}, inplace=True)
df.rename(columns = {'contact':'contact type'}, inplace=True)
df.rename(columns = {'campaign':'total contacts'}, inplace=True)
df.rename(columns = {'emp var rate':'employment variation rate'}, inplace=True)
df
```

### Capitalize column strings

```python
columns = df.columns
columns = [c.title() for c in columns]
columns
```

```python
df.columns = columns
df
```

### Clean data by column

```python
df['Contact Duration'].value_counts()
```

```python
rgx = re.compile(r'\d+(y)')
df['Job Type'] = df['Job Type'].str.replace('.', '')
df['Job Type'] = df['Job Type'].str.replace('-', ' ')
df['Education'] = df['Education'].str.replace('.', ' ')
df['Education'] = df['Education'].str.replace('university degree', 'college')
df['Education'] = df['Education'].str.replace('y', 'th')
df
```

### Transpose the DataFrame to see all column names inline

```python
df.T
```

### Convert Days Since Contacted data to int64.

```python
df['Days Since Contacted'].replace(999, 99999, inplace=True)
```

### Change 999 to 99999 to allow a very large number to represent not contacted
#### NOTE: Converting to NaN casts the column as float64 by default, but int64 isa  more appropriate data type.

```python
df['Mortgage'].replace('unknown', np.nan, inplace=True)
df['Personal Loan'].replace('unknown', np.nan, inplace=True)
df['Previous Outcome'].replace('nonexistent', np.nan, inplace=True)
df
```

```python
df.dtypes
```

### Check for missing values in numeric fields

```python
np.count_nonzero(df[['Consumer Price Index',
                     'Consumer Confidence Index',
                     'Eurobori 3 Mos',
                     'Employment Variation Rate',
                     'Number Employed',
                     'Days Since Contacted',
                     'Total Contacts',
                     'Contact Duration',
                     'Number Previous Contacts']].isnull())
```

# Conclusion
### This notebook cleaned a dataset with columns and data points that had unconventional and obfuscated formatting. Variable names were renamed and string operations were performed on both variables and values to provide clarity to anyone attempting to perform analysis on the data. Any "unknown" values were left as is if the data type of that variable was an object. Missing data in columns with numeric data types were converted to NaN using np.nan. The only exception to this rule was the Days Since Contacted column, which was more appropriately cast as an int64 datatype, and converting 999 to NaN using numpy automatically recast the column as float64.
